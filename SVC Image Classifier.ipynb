```python
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np
import os
import cv2
import pickle

```

```python
# Load the dataset and labels
dataset = []  # List of feature vectors extracted from images
labels = []   # List of associated labels

for folder in os.listdir('after'):
    for image in os.listdir(f'after/{folder}'):
        img = cv2.imread(f'after/{folder}/{image}')
        dataset.append(img)
        labels.append(folder)
```

```python



# Prepare the dataset for training
X = np.array(dataset)  # Convert the dataset to a numpy array
y = np.array(labels)   # Convert the labels to a numpy array



```

```python
X
```

```python
y
```

```python
X = np.array([image.flatten() for image in dataset])  # Flatten each image to a 1D array
y = np.array(labels)   # Convert the labels to a numpy array
```

```python
len(X[0])
```

```python
# Create an SVM classifier
classifier = svm.SVC()

# Train the SVM classifier
classifier.fit(X, y)
```

```python

# save the model to disk
filename = 'PLATE_model.sav'
pickle.dump(classifier, open(filename, 'wb'))
```

```python
total_accuracy = []
predict = []
for i in range( len( os.listdir('extract_test') ) ):
    # Load the dataset and labels
    dataset_test = []  # List of feature vectors extracted from images
    labels_test = []   # List of associated labels

    folder = f'datatest{i+1}'

    if os.path.exists( f'extract_test/{folder}') :
        for image in os.listdir(f'extract_test/{folder}'):
            img = cv2.imread(f'extract_test/{folder}/{image}')
            img = cv2.resize(img,(100,200))
            dataset_test.append(img)
            basename, extension = os.path.splitext(image)
            labels_test.append(basename.split('.')[1])

        X_test = np.array(dataset_test)  # Convert the dataset to a numpy array
        y_test = np.array(labels_test)
        X_test = np.array([image.flatten() for image in dataset_test])

        loaded_model = pickle.load(open('PLATE_model.sav', 'rb'))
        # Make predictions on the testing set
        y_pred = loaded_model.predict(X_test)

        # Evaluate the performance of the SVM classifier
        accuracy = accuracy_score(y_test, y_pred)
        predict.append(" ".join([str(item) for item in y_pred]))
        print(f'\nnama: {folder}\nprediction: {y_pred} \nactual data: {y_test}\nAccuracy: {accuracy}')
        total_accuracy.append(accuracy)
    else:
        print(f'\nnama: {folder}\nprediction: {0} \nactual data: {0}\nAccuracy: {0}')
   
```
